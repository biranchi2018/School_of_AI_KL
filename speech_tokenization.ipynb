```
def get_label(string):
  strings = string.split('/')
  left_string = ' '.join(strings[0].split('-')[:-1])
  right_string = strings[1].replace('.wav', '')
  return left_string + ' ' + right_string
```

```
labels = [get_label(s) for s in sebut_perkataan_man]
labels[:10]
```

```
# link to augmentation notebook, https://github.com/huseinzol05/Bahasa-NLP-Tensorflow/blob/master/speech-to-text/augmentation.ipynb
```

```
vocab_word = list(set(' '.join(labels).split(' ')))
vocab_word[:10]
```

```
len(vocab_word)
```

```
with open('word-corpus.txt', 'w') as fopen:
  fopen.write('\n'.join(vocab_word))
```

```
import youtokentome as yttm

model_path = "example.model"
yttm.BPE.train(data='word-corpus.txt', 
               vocab_size=200, model=model_path)
```

```
bpe = yttm.BPE(model=model_path)
```

```
bpe.encode(['Biranchi'], output_type=yttm.OutputType.SUBWORD)
```

```
encode_id = bpe.encode(['Biranchi'], output_type=yttm.OutputType.ID)
encode_id
```

```
bpe.decode(encode_id)
```
